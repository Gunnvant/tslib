### ADIDA Model

```python
import pandas as pd
import os
os.chdir(os.path.expanduser(".."))
import tslib
import warnings
```

```python
warnings.filterwarnings('ignore') 
```

```python
df = pd.read_csv("./data/daily_min_temp.csv")
```

```python
from tslib import models
```

```python
from statsforecast.utils import ConformalIntervals
```

```python
m = models.ADIDAModel(prediction_intervals=ConformalIntervals())
```

```python
m = m.fit(ts=df.Temp)
```

```python
m.forecast(steps=3)
```

```python
m.residual_analysis()
```

```python
m.plot_fit()
```

```python
m.get_fit_interval()
```

```python
m.get_forecast_interval()
```
