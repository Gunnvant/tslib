## ADIDA CV Examples

```python
import pandas as pd
import os
os.chdir(os.path.expanduser(".."))
import tslib
import warnings
warnings.filterwarnings('ignore') 
```

```python
df = pd.read_csv("./data/daily_min_temp.csv")
```

```python
from tslib import preprocessing,evaluation
```

```python
from tslib import models
```

```python
from statsforecast.utils import ConformalIntervals
m = models.ADIDAModel(prediction_intervals=ConformalIntervals())
```

```python
m = m.fit(ts=df.Temp)
```

```python
proc = preprocessing.CVPreprocessor(start_idx=2000,step_size=180,horizon=365)
```

```python
eval = evaluation.CrossValidation()
```

```python
folds = proc.fit(df.Temp)
```

```python
r = eval.evaluate(folds,m,type='adida')
```

```python
r1 = r['mae']
r1
```
