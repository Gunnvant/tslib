### Prophet Model

```python
import pandas as pd
import os
os.chdir(os.path.expanduser(".."))
import tslib
```

```python
df = pd.read_csv("./data/daily_min_temp.csv")
```

```python
from tslib import models
```

```python
ts = df.rename(columns={'Date':'ds','Temp':'y'})
```

```python
p = models.ProphetModel()
```

```python
p.fit(ts)
```

```python
p.forecast(steps=2)
```

```python
p.summary()
```

```python
p.residual_analysis()
```

```python
p.plot_fit()
```

```python
p.get_fit_interval()
```

```python
p.get_forecast_interval()
```
